```python
import numpy as np
import matplotlib.pyplot as plot

n = np.arange(-20, 21, 1)
```

### ${\left(\frac{9}{10}\right)^n}u[n]$

```python
x_n = np.power((9 / 10), n)
plot.stem(n, x_n)
plot.show()
```

### $3 \cos \left[ \frac{\pi n}{4} + \frac{\pi}{3} \right]$

```python
x_n = np.cos((np.pi / 4) * n + np.pi / 3)
plot.stem(n, x_n)
plot.show()
```

### $(-0.75)^n u[n]$


```python
def u(n):
    x = np.array(n)
    x[(x >= 0)] = 1
    x[~(x >= 0)] = 0
    return x
```

```python
x_n = np.power((-.75), n) * u(n)
plot.stem(n, x_n)
plot.show()
```

### $\frac{\sin (\pi n/6 ) }{\pi n/6}$

Please make sure that the value at n=0 is not undefined. Consider the limit value

```python
def sinc(x):
    np.seterr(invalid='ignore')
    x_n = np.sin(np.pi * x) / (np.pi * x)
    x_n[np.isnan(x_n)] = 1
    return x_n
```

```python
x_n = sinc(n / 6)
plot.stem(n, x_n)
plot.show()
```

```python
# Using numpy sinc 
x_n = np.sinc(n / 6)
plot.stem(n, x_n)
plot.show()
```

### $e^{-0.1 n} \cos\left( \frac{\pi n}{8} \right)$

```python
x_n = np.exp(-.1 * n) * np.cos(np.pi / 8 * n)
plot.stem(n, x_n)
plot.show()
```

### ${\sin (\pi n^2 / 200 ) }$

```python
x_n = np.sin((np.pi * (n**2)) / 200)
plot.stem(n, x_n)
plot.show()
```
