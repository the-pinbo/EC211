 <li>$ \frac{1}{(1+az^{-1})^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Single pole)  </li>
  <li>$\frac{1}{1-az^{-1}+{(az^{-1})}^2} \,\, for \,\, a = \pm{0.8},\pm{1},\pm{2}$ (Complex conjugate poles) </li>    
</ol>

https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.signal.tf2zpk.html

```python
def unitstep(n):
    if (n >= 0):
        return 1
    else:
        return 0
n = np.arange(-2,30)
u = np.array([unitstep(i) for i in n])
```

Q5. For the LTI systems described by the following difference equation. Obtain the pole zero plot. Comment on the type of response.

$y[n] + 1.3y[n - 1] + 1.29y[n - 2] + 0.52y[n - 3] + 0.16y[n - 4] = x[n] - 0.9x[n - 1]$

$y[n] - 1.3y[n - 1] + 1.29y[n - 2] - 0.52y[n - 3] + 0.16y[n - 4] = x[n] + 0.9x[n - 1]$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
#Y(Z)(1 + 1.3Z^-1 + 1.29Z^-2 + 0.52Z^-3 + 0.16Z^-4) = X(Z)(1 - 0.9Z^-1)

b=[1,-0.9]
a=[1,1.3,1.29,0.52,0.16]
z,p,k =sig.tf2zpk(b,a)
print(z,p)

plt.figure(figsize=(5, 5))
plt.plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
n1 = np.arange(0,np.pi*2,1/100)
plt.plot(np.sin(n1),np.cos(n1))
plt.show()
```

```python
#Y(Z)(1 - 1.3Z^-1 + 1.29Z^-2 - 0.52Z^-3 + 0.16Z^-4) = X(Z)(1 + 0.9Z^-1)

b=[1,0.9]
a=[1,-1.3,1.29,-0.52,0.16]
z,p,k =sig.tf2zpk(b,a)
print(z,p)

plt.figure(figsize=(5, 5))
plt.plot(z.real, z.imag, 'ro', p.real, p.imag, 'rx')
n1 = np.arange(0,np.pi*2,1/100)
plt.plot(np.sin(n1),np.cos(n1))
plt.show()
```
