Q4. Solve the system with difference equation $y[n] - 0.8y[n - 1] = 0.5(x[n] + x[n - 1])$ with the initial condition is given by $y[-1] = -2$. Plot the output sequence for input 
$~~ (a)~ x[n] = u[n] - u[n - 10] $ and  $~~ (b)~ x[n] = 2(u[n] - u[n - 10])$. Could the system be linear? Repeat the experiment
with zero initial conditions. ( Use scipy.sig.lfilter )

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
n = np.arange(-2,30)
u = np.array([unitstep(i) for i in n])
u10 = np.array([unitstep(i-10) for i in n])
input= u - u10

#(a) with initial condition

#Y(Z) - 0.8(Z^-1Y(Z) + y(-1)) = 0.5(X(z) + Z^-1X(z) + x(-1)) but x(-1) = 0
#Y(Z) - 0.8(Z^-1Y(Z) -2) = 0.5(X(z) + Z^-1X(z))
#Y(Z) - 0.8y[n-1] + 1.6delta(n) = 0.5(x[n] + x[n-1])

b1=[0.5,0.5]
a1=[1.0,-0.8]
responseA = sig.lfilter(b1,a1,input) #input applied on signal 3
responseA=list(responseA)
responseA[2] = responseA[2] - 1.6  #need to add 1.6 delta , TO TAKE CARE OF INITIAL CONDITIONS
plt.stem(n,responseA,use_line_collection=True)
plt.title("x[n]=u[n]−u[n−10] initial condition is given by  y[−1]=−2 ")
plt.show(responseA)

# (b) with initial condition

responseB = sig.lfilter(b1,a1,2*input) #input applied on signal 3
responseB=list(responseB)
responseB[2] = responseB[2] - 1.6  #need to add 1.6 delta , TO TAKE CARE OF INITIAL CONDITIONS
plt.stem(n,responseB,use_line_collection=True)
plt.title("x[n]= 2(u[n]−u[n−10]) initial condition is given by  y[−1]=−2 ")
plt.show(responseB)


#(a) WITHHOUT initial condition

#Y(Z) - 0.8(Z^-1Y(Z) + y(-1)) = 0.5(X(z) + Z^-1X(z) + x(-1)) but x(-1) = 0
#Y(Z) - 0.8(Z^-1Y(Z) -2) = 0.5(X(z) + Z^-1X(z))
#Y(Z) - 0.8y[n-1] + 1.6delta(n) = 0.5(x[n] + x[n-1])

b1=[0.5,0.5]
a1=[1.0,-0.8]
responseC = sig.lfilter(b1,a1,input) #input applied on signal 3
responseC=list(responseC)
plt.stem(n,responseC,use_line_collection=True)
plt.title("x[n]=u[n]−u[n−10]")
plt.show(responseC)

#(b) WITHOUT initial condition

responseD = sig.lfilter(b1,a1,2*input) #input applied on signal 3
responseD=list(responseD)
plt.stem(n,responseD,use_line_collection=True)
plt.title("x[n]= 2(u[n]−u[n−10])")
plt.show(responseD)
```
