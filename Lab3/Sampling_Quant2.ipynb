Q1. Generate the following signals with the specified sampling rate and listen to it.   Let $t$ vary from $0 - 3$seconds. Let the sampling rate is 8000Hz or a sampling time period is $125 \mu s$  (A sample is taken at every integer multiple of $125 \mu s$ )
 <ol>
  <li>$x(t) = sin(100\pi t)$ (100 Hz sinusoid)  </li>
 <li>$x(t) = cos(1000\pi t)$ (1kHz sinusoid)  </li>
  <li>$x(t) = sin(100 \pi t^2) $  (Called a chirp signal) </li>
</ol>
To listen to a signal, write the signal to a file using wave write command  from scipy
Please look up the documentation here (https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.io.wavfile.write.html).   Then download the signal and listen to it. 

```python
import numpy as np 
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt

# Set the sampling rate
rate = 8000;

# Genearate the t values use np.arange
t = np.arange(0,3,(1.0/8000))

#Generate signal 
s1 = np.sin(100*np.pi*t)
plt.plot(s1)
s2 = np.cos(1000*np.pi*t)
s3 = np.sin(100*np.pi*t*t)
#Write thee signal 
wav.write('sin100.wav',rate, s1);
wav.write('cos1000.wav',rate, s2);
wav.write('sin100t^2.wav',rate, s3);



```

Q3. In DTMF dialling a number is represented by a dual frequency tone. Do a
web search and find the frequencies of each digit. Generate DTMF tones
corresponding to the telephone number 08242474040 by sampling the sum
of sinusoids at the required frequencies at Fs = 8192 Hz. Concatenate the
signals by putting 100 zeros between each signal (to represent silence) and
listen to the signal. (Must sound like tone dialling the number from
a phone)

```python
import numpy as np 
import resampy
import scipy.io.wavfile as wav
rate=8192

#DTMF Tone correspoding to 0
sample0 , tone0 = wav.read('DTMF-0.wav')
resampy.core.resample(tone0,48000,8192)
# DTMF Tone correspoding to 8
sample8 , tone8 = wav.read('DTMF-8.wav')
resampy.core.resample(tone8,48000,8192)
#DTMF Tone correspoding to 2
sample2 , tone2 = wav.read('DTMF-2.wav')
resampy.core.resample(tone2,48000,8192)
#DTMF Tone correspoding to 4
sample4 , tone4 = wav.read('DTMF-4.wav')
resampy.core.resample(tone4,48000,8192)
#DTMF Tone correspoding to 7
sample7 , tone7 = wav.read('DTMF-7.wav')
resampy.core.resample(tone7,48000,8192)

tone = np.concatenate((tone0,tone8,tone2,tone4,tone2,tone4,tone7,tone4,tone0,tone4,tone0))

wav.write('tone.wav',rate, tone);

plt.figure(figsize=(100,20))
plt.xlim(0,400000)
plt.plot(tone)
```
